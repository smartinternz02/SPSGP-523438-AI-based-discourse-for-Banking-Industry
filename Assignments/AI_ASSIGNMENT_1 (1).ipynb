QUESTION 1: Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)

```python
import numpy as np
import pandas as pd
#generating random values
d = np.random.rand(4, 4)
#creating a DataFrame 
df = pd.DataFrame(d, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'], index=['Observation 1', 'Observation 2', 'Observation 3', 'Observation 4'])
#print the DataFrame
print(df)   
```

QUESTION 2: Rename the task - 1 'df' dataframe column names to 'Random value 1', 'Random value 2', 'Random value 3' & 'Random value 4'

```python
import numpy as np
import pandas as pd
#renaming columns
df = df.rename(columns = {'Feature 1' : 'Random value 1',
                          'Feature 2' : 'Random value 2',
                          'Feature 3' : 'Random value 3',
                          'Feature 4' : 'Random value 4'})
#print the DataFrame
print(df)
```

QUESTION 3: Find the descriptive statistics of the 'df' dataframe.

```python
import numpy as np
import pandas as pd
#statistical description
stat = df.describe()
#print the Dataframe
print(df)
print('\n')
print(stat)
```

QUESTION 4: Check for the null values in 'df' and find the data type of the columns.

```python
import numpy as np
import pandas as pd
#checking the null values
null_val = df.isnull().sum()
#checking data types
data_type = df.dtypes
#print the DataFrame
print(null_val)
print('\n')
print(data_type)
```

QUESTION 5: Display the 'Random value 2' & 'Random value 3' columns with location method and index location method.

```python
import numpy as np
import pandas as pd
#display columns using the location method
location_method = df[['Random value 2', 'Random value 3']]
print("Columns using location method:")
print(location_method)
#display columns using the index location method
index_location_method = df.iloc[:, 1:3]
print("\nColumns using index location method:")
print(index_location_method)
```
